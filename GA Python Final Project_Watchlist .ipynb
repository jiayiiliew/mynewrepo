```python
#if we want to classify the movies based on their IMDb Ratings, we can define a function to do so and then apply the function to the dataframe as shown below
```

```python
# Classify movies based on IMDb Ratings
def rating_group(rating):
    if rating >= 7.5:
        return 'Good'
    elif rating >= 6.0:
        return 'Average'
    else:
        return 'Bad'
```

```python
# Lets apply this function on our movies data
# creating a new variable in the dataset to hold the rating category
watchlist_df['Rating_category'] = watchlist_df['IMDb Rating'].apply(rating_group)
```

```python
watchlist_df[['Title','Directors','IMDb Rating','Rating_category']].head(5)
```
